# Question 7

Question 7: You are given a customer review:
review = "The product quality is Excellent but delivery was Late"
Write Python code to:
1. Convert to lowercase
2. Count number of words
3. Check if the word "excellent" exists
4. Replace "Late" with "delayed"

````python
review = "The product quality is Excellent but delivery was Late"

review_lower = review.lower()
word_count = len(review.split())
contains_excellent = "excellent" in review_lower
updated_review = review.replace("Late", "delayed")

print("Lowercase Review:", review_lower)
print("Word Count:", word_count)
print("Contains 'excellent':", contains_excellent)
print("Updated Review:", updated_review)
````

# Question 8

Question 8: Given sales data:
sales = [12000, 15000, 11000, 18000, 17000]
1. Calculate total revenue.
2. Find average revenue.
3. Identify highest sales month.
4. Append new month sale = 20000.
5. Sort the list in descending order.

````python
sales = [12000, 15000, 11000, 18000, 17000]

total_revenue = sum(sales)
average_revenue = total_revenue / len(sales)
highest_sale = max(sales)

sales.append(20000)
sales.sort(reverse=True)

print("Total Revenue:", total_revenue)
print("Average Revenue:", average_revenue)
print("Highest Sale:", highest_sale)
print("Updated Sales List:", sales)
````

# Question 9

You are building a simple sales analysis system.
Given:


```
# products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]
```


Tasks:
1. Compute total revenue for each product.
2. Identify which product generated maximum revenue.
3. Store revenues in a new list.
4. Print a formatted summary like:

````python
products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]

revenues = []

for i in range(len(products)):
    revenues.append(prices[i] * quantities[i])

max_revenue = max(revenues)
max_product = products[revenues.index(max_revenue)]

print("Sales Summary")
print("---------------------------")

for i in range(len(products)):
    print(
        "Product:", products[i],
        "| Revenue: ₹", revenues[i]
    )

print("\nHighest Revenue Product:", max_product)
print("Revenue Generated: ₹", max_revenue)
````
